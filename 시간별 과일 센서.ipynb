# 감


```python
import pandas as pd
import os
from io import StringIO

# The directory where your text files are stored
directory = "C:\\Users\\lee\\Desktop\\각 과일별 나열\\감"

# Function to read the contents of each file into a dataframe
def read_file_to_df(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        data_str = file.readlines()[2:-1]  # Skipping the first two and the last line as per your example
        data_combined = ''.join(data_str)
        data_io = StringIO(data_combined)
        return pd.read_csv(data_io, header=None, sep=',')

# Initialize an empty dataframe to store all the data
all_data = pd.DataFrame()

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.txt'):  # Check if the file is a .txt file
        filepath = os.path.join(directory, filename)
        df = read_file_to_df(filepath)
        all_data = all_data.append(df, ignore_index=True)

# Assuming all files have the same structure and columns as per your example
all_data.columns = ['Time', 'MQ-2', 'MQ-3', 'MQ-136', 'MQ-137']

# Now you can perform further analysis or save the combined dataframe
# For example, to save as a CSV:
all_data.to_csv('C:\\Users\\lee\\Desktop\\combined_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Drop the 'Time' column
all_data.drop('Time', axis=1, inplace=True)

# Plotting the remaining columns
plt.figure(figsize=(10, 5))
for column in all_data.columns:
    plt.plot(all_data.index, all_data[column], label=column)

plt.title('Sensor Readings Over Time')
plt.xlabel('Index')
plt.ylabel('Sensor Readings')
plt.legend()
plt.grid(True)
plt.show()
```
